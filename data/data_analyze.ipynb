```python
# 형식적인 말 삭제

crawling_df['processed_reviews'] = crawling_df['full_text'].copy()

remove_text = [
    '*사진 필수 게시판입니다. 꼭 사진 첨부를 해주세요 제목양식: 업체명 + 후기내용 ex) mmw스튜디오 + 계약후기에요! / mmw한복 + 본식후기입니다 \u200b'
]

crawling_df['processed_reviews'] = crawling_df['processed_reviews'].str.replace(remove_text[0], '', regex=False).str.strip()

exceptions_df = crawling_df[crawling_df['processed_reviews'].str.contains(remove_text[0], regex=False, na=False)]

print("--- 삭제되지 않고 남은 full_text 목록 ---")
if not exceptions_df.empty:
    print(exceptions_df['full_text'].tolist())
else:
    print("모든 문구가 성공적으로 제거되었습니다.")
```

```python
# 1. 검사할 단어 리스트 설정
check_list = [
    '1. 베뉴 이름', 
    '2. 견적받은 날짜', 
    '3. 직접 찍은 사진 첨부 필수 (공사중이거나, 코로나 때문에 못 첨부할시에는 꼭 글로 명시하시고 홈페이지 사진 대체)',
    '3. 직접 찍은 사진 첨부 필수',
    '4. 정가 식대, 대관료 그 외 기타 비용',
    '5. 견적받은 할인가 식대, 대관료 그 외 기타 비용',
    '6. 선택하게 된 계기',
    '7. 아쉬운 점 혹은 걱정되는 점',
    '8. 여러 업체의 정보를 올릴 경우 필히 한 글에 묶어서 올려주세요.'
    
]

# 2. 누락된 데이터 확인용 리스트
missing_rows = []

# 3. 로우별로 순회하며 확인
for idx, row in crawling_df.iterrows():
    text = str(row['full_text']) # 검사할 컬럼명 확인 필수
    missing_words = [word for word in check_list if word not in text]
    
    if missing_words:
        # 인덱스 번호와 누락된 단어들을 저장
        missing_rows.append({
            'row_number': idx,
            'missing_content': missing_words
        })

# 4. 결과 출력
if missing_rows:
    print(f"⚠️ 총 {len(missing_rows)}개의 행에서 필수 항목이 누락되었습니다.\n")
    for item in missing_rows:
        print(f"📍 행 번호 {item['row_number']}: 누락 항목 -> {item['missing_content']}")
else:
    print("✅ 모든 행에 필수 항목이 포함되어 있습니다.")
```

```python
import pandas as pd
import re

# 1. 추출할 태그 리스트 (영문 컬럼명)
tags = [
    'name', 'hall_vibe', 'catering', 'parking_facility', 'transport_access', 
    'bridal_room', 'guest_flow', 'pricing_detail', 'service_etc', 'disadvantage', 'in_guarantee'
]

# column = ['hall_name_raw', 'full_text', 'gpt-processed']
# crawling_df = pd.read_csv(r'C:\Users\yiuri\OneDrive\GitHub\Recsys-Wedding\data\20251230_141950_gpt_results.csv',header=None, names=column)
crawling_df = pd.read_csv(r'C:\Users\yiuri\OneDrive\GitHub\Recsys-Wedding\data\20251230_141950_gpt_results.csv')


def parse_gpt_tags(text):
    text = str(text)
    extracted_data = {}
    
    for tag in tags:
        # 정규표현식 설명:
        # <tag> : 시작 태그를 찾음
        # (.*?) : 다음 태그(<)가 나오거나 문자열 끝($)이 나올 때까지의 내용을 캡처
        # (?=<|$) : 전방 탐색을 통해 다음 태그 직전까지만 가져옴
        pattern = rf"<{tag}>(.*?)(?=<|$)"
        match = re.search(pattern, text, re.DOTALL)
        
        if match:
            extracted_data[tag] = match.group(1).strip()
        else:
            extracted_data[tag] = "None"
            
    return pd.Series(extracted_data)


tag_df = crawling_df['gpt-processed'].apply(parse_gpt_tags)


final_df = pd.concat([crawling_df, tag_df], axis=1)


print(tag_df.head(2))
tag_df.to_csv('final_weddingDatasets.csv', index=True, encoding='utf-8-sig')
```

```python
import re
import pandas as pd

def string_to_int(row):
    # row에는 인덱스와 값이 포함됩니다.
    idx = row.name
    text = str(row['in_guarantee'])
    
    # 1. 숫자만 추출 (정규표현식 사용)
    # text에서 숫자가 아닌 모든 문자를 제거합니다.
    clean_text = re.sub(r'[^0-9]', '', text)
    
    # 2. 숫자 외의 글자가 포함되어 있었는지 체크 (출력용)
    if clean_text != text:
        print(f"⚠️ [Index {idx}] 글자 포함됨: '{text}' -> '{clean_text}'로 변환")
    
    # 3. 빈 값인 경우 처리 (None이나 빈 문자열이면 0으로)
    if not clean_text:
        return 0
        
    # 4. int로 변환하여 반환
    return int(clean_text)

# 적용하기 (axis=1을 써야 row.name으로 인덱스 접근이 가능합니다)
tag_df['in_guarantee'] = tag_df.apply(string_to_int, axis=1)

# 결과 확인
print("\n✅ 변환 완료:")
print(tag_df['in_guarantee'].head())
```
